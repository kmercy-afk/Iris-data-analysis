```python
from sklearn.datasets import load_iris
import pandas as pd

# Load the iris dataset
iris = load_iris()

# Create DataFrame for the features
X = pd.DataFrame(iris.data, columns=iris.feature_names)

# Create DataFrame for the target variable (Species)
y = pd.DataFrame(iris.target, columns=["Species"])

# Combine X and y into a single DataFrame
Xy = pd.concat([X, y], axis=1)

# 1. Display the first four samples and see what each feature is
print("First 4 samples:")
print(Xy.head(4))

# 2. Output the total number of samples for each label (species)
print("\nTotal number of samples for each label (species):")
print(y['Species'].value_counts())

# 3. Check if features have missing values
print("\nCheck for missing values in features:")
print(Xy.isnull().sum())

# 4. Display the mean, standard deviation, and quartiles of a feature
print("\nDescriptive statistics for the 'sepal length (cm)' feature:")
print(Xy['sepal length (cm)'].describe())
```

```python
# General information about the dataset
print("\nGeneral information about the dataset:")
print(Xy.info())

# Descriptive statistics for all numerical features
print("\nDescriptive statistics for all numerical features:")
print(Xy.describe())
```

```python
# Using .loc[] to extract a specific column
sepal_width_2 = Xy.loc[:, 'sepal width (cm)']
print(sepal_width_2.head())
```

```python
# Extracting rows 50 to 99 of the 'petal length (cm)' column
petal_length = Xy.loc[50:99, 'petal length (cm)']
print(petal_length)
```

```python
# Extracting rows 50 to 99 of the 'petal width (cm)' column
petal_width = Xy.loc[50:99, 'petal width (cm)']
print(petal_width)
```

```python
# Extracting rows where petal_width is 0.2
petal_width_02 = Xy.loc[Xy['petal width (cm)'] == 0.2]
print(petal_width_02)
```

.loc[]:

.loc[] is label-based indexing, meaning that it uses the labels of the rows and columns.
It allows you to select both rows and columns using row/column labels (names).
It is inclusive of both the start and stop labels when selecting a range of rows.
Example: Xy.loc[50:99, 'petal length (cm)'] will select rows 50 through 99 (inclusive) and the specified column.
.iloc[]:

.iloc[] is position-based indexing, meaning that it uses integer positions (indices) to select rows and columns.
You specify the row and column positions as integers, starting from 0.
It is exclusive of the stop position when selecting a range of rows or columns (similar to Python’s list slicing).
